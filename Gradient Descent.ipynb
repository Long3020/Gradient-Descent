```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):# Hàm f(x)
    return np.sin(0.5 * x) + 0.1 * x**2 - 0.5 * np.exp(-0.1 * x)
```

```python
def df(x):  # Đạo hàm của f(x)
    return 0.5 * np.cos(0.5 * x) + 0.2 * x + 0.05 * np.exp(-0.1 * x)
```

```python
# Gradient Descent
def gd(x0 = 4, eta = 1, iterations = 100):
    x = [x0]
    for it in range(iterations):
        x_new = x[-1] - eta*df(x[-1]) # Công thức GD
        x.append(x_new)
    return x
```

```python
# Chạy gradient descent
x_vals = gd(x0=4, eta=1, iterations=100)
y_vals = [f(x) for x in x_vals]

# Vẽ đồ thị
x_plot = np.linspace(-10, 6, 100)
y_plot = f(x_plot)

plt.figure(figsize=(8, 6))
plt.plot(x_plot, y_plot, label='f(x)', color='blue')
plt.scatter(x_vals, y_vals, color='red', label='Gradient Descent Steps')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.title('Gradient Descent Optimization')
plt.show()
```
